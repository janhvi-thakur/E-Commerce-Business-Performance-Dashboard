```python
pareto = (
    df.groupby('Product Name')['Sales']
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

pareto['Cumulative Sales'] = pareto['Sales'].cumsum()

pareto['Cumulative %'] = (
    pareto['Cumulative Sales'] /
    pareto['Sales'].sum()
) * 100

pareto.head(20)
```

```python
top_80 = pareto[pareto['Cumulative %'] <= 80]

print("Products contributing to 80% of sales:", len(top_80))
```

# Executive Summary

This report analyzes four years (2015–2018) of e-commerce sales data to identify key trends, customer behavior, regional performance, and product performance. The objective is to provide actionable insights that support strategic business decisions.

# Executive Business Insights

## Insight 1: Category Performance

**Observation:**
The Technology category generated the highest sales among all product categories.

**Business Impact:**
Technology products are the primary revenue drivers for the business and contribute significantly to overall sales.

**Recommendation:**
The company should continue investing in the Technology category by expanding product offerings, maintaining sufficient inventory, and running targeted marketing campaigns to maximize revenue.

---

## Insight 2: Regional Performance

**Observation:**
The West region recorded the highest sales compared to the other regions.

**Business Impact:**
The West region represents the strongest market and contributes significantly to overall business revenue.

**Recommendation:**
Maintain strong customer engagement in the West while implementing region-specific marketing strategies to improve sales in lower-performing regions.

---

## Insight 3: Customer Segment Analysis

**Observation:**
The Consumer segment contributed the largest share of total sales.

**Business Impact:**
A large portion of the company's revenue depends on Consumer customers.

**Recommendation:**
Continue strengthening customer retention programs for the Consumer segment while developing targeted campaigns for the Home Office and Corporate segments to diversify revenue sources.

---

## Insight 4: Customer Analysis

**Observation:**
A small group of customers contributes a significant portion of the company's total sales.

**Business Impact:**
Retaining these high-value customers is essential for maintaining business growth.

**Recommendation:**
Introduce loyalty programs, personalized discounts, and exclusive offers for high-value customers to increase customer lifetime value.

---

## Insight 5: Product Performance

**Observation:**
A limited number of products generated a large proportion of total revenue.

**Business Impact:**
These products are key revenue generators and have a direct impact on overall business performance.

**Recommendation:**
Ensure these products are consistently available in inventory and prioritize them in promotional campaigns.

---

## Insight 6: Sales Trend

**Observation:**
Sales generally increased over the four-year period from 2015 to 2018, indicating positive business growth.

**Business Impact:**
The company has experienced consistent revenue growth and increasing customer demand.

**Recommendation:**
Continue investing in successful sales strategies while identifying factors that contributed to growth for future planning.

---

## Insight 7: Seasonal Analysis

**Observation:**
Certain months generated significantly higher sales than others, indicating seasonal purchasing behavior.

**Business Impact:**
Customer demand varies throughout the year, affecting inventory and marketing requirements.

**Recommendation:**
Increase inventory levels and marketing activities before peak sales months while planning promotional campaigns during slower periods.

---

## Insight 8: Pareto Analysis

**Observation:**
Approximately 20% of the products contributed to nearly 80% of total sales.

**Business Impact:**
A relatively small number of products drive the majority of business revenue.

**Recommendation:**
Prioritize inventory management, pricing strategies, and promotional efforts for these high-performing products while reviewing low-performing products for improvement or discontinuation.

---

## Overall Conclusion

The analysis indicates steady business growth between 2015 and 2018, with Technology products, Consumer customers, and the West region being the primary contributors to revenue. Strategic investments in high-performing products and customer segments, combined with targeted initiatives for underperforming regions and segments, can further improve business performance and support sustainable long-term growth.
